## Date vs Total Demand   
A graphical representation of date vs total demand of energy. This information is quite useful as it shows which times of the year you'd expect to see a high demand for energy.

```python
plt.figure()
ax = energy_data.plot(x='Date',y='TotalDemand')
plt.xlabel('Date')
plt.ylabel('TotalDemand')
plt.xticks(rotation=45)
plt.show()
```

## Air vs Sea and Station Pressure
A graphical representation of air vs sea and station pressure. The comparison between air temperature and pressure over a number of years can be helpful in predicting certain weather patterns as pressure plays a role in the determining the temperature.

```python
plt.figure()
ax = temperature_data.plot(x='Date',y='StationPressure',color='red')
temperature_data.plot(x='Date',y='SeaPressure',ax=ax,color='blue')
plt.ylabel('Hectopascals (Hpa)')
plt.ylim(0,1500)
plt.xlabel('Date and Time')
plt.xticks(rotation=45)
plt.legend('')

ax2 = ax.twinx()
temperature_data.plot(x='Date',y='AirTemperature',ax=ax2,color='green',legend='None')
plt.ylabel('Degrees Celsius')
plt.ylim(0,60)
plt.legend('')
plt.title('StationPressure (red), SeaPressure (blue) and AirTemperature (green) temperatures')
plt.show()
```

# Single day comparision of temperature and energy demand

These graphs show single day comparisions of energy demand (TotalDemand column) between regions.
Generally the energy demand from New South Wales is higher than that in other regions and Tasmania & South Australia are far lower than other regions.
There are always two periods of increased energy demand, which tend to happen at around the same time in all states.
This patter persists through time (seen by the graphs of 2015 having the same overall shape as those from 2005).

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))

data.groupby("Region", sort=False).get_group("NSW")["2005-06-01"]["TotalDemand"].plot(ax=axes[0,0])
data.groupby("Region").get_group("VIC")["2005-06-01"]["TotalDemand"].plot(ax=axes[0,1])
data.groupby("Region").get_group("QLD")["2005-06-01"]["TotalDemand"].plot(ax=axes[0,2])
data.groupby("Region").get_group("TAS")["2005-06-01"]["TotalDemand"].plot(ax=axes[1,0])
data.groupby("Region").get_group("SA")["2005-06-01"]["TotalDemand"].plot(ax=axes[1,1])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))

data.groupby("Region", sort=False).get_group("NSW")["2015-06-01"]["TotalDemand"].plot(ax=axes[0,0])
data.groupby("Region").get_group("VIC")["2015-06-01"]["TotalDemand"].plot(ax=axes[0,1])
data.groupby("Region").get_group("QLD")["2015-06-01"]["TotalDemand"].plot(ax=axes[0,2])
data.groupby("Region").get_group("TAS")["2015-06-01"]["TotalDemand"].plot(ax=axes[1,0])
data.groupby("Region").get_group("SA")["2015-06-01"]["TotalDemand"].plot(ax=axes[1,1])
```

These graphs show that there is a strong correlation between temperature and energy demand!
We know this as the orange energy curve seems to increase in value with the blue temperature curve.
Often though the highest temperature peaks are not accompanied by increased energy usage.
We once again graph a day from two years (2005 and 2018) in every region to ensure we don't describe a unique pattern which does not repeat itself in majority of the days.

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))

data.groupby("Region", sort=False).get_group("NSW")["2005-06-01"]["WetBulbTemperature"].plot(ax=axes[0,0])
data.groupby("Region", sort=False).get_group("NSW")["2005-06-01"]["TotalDemand"].plot(ax=axes[0,0], secondary_y=True)
data.groupby("Region").get_group("VIC")["2005-06-01"]["WetBulbTemperature"].plot(ax=axes[0,1])
data.groupby("Region").get_group("VIC")["2005-06-01"]["TotalDemand"].plot(ax=axes[0,1], secondary_y=True)
data.groupby("Region").get_group("QLD")["2005-06-01"]["WetBulbTemperature"].plot(ax=axes[0,2])
data.groupby("Region").get_group("QLD")["2005-06-01"]["TotalDemand"].plot(ax=axes[0,2], secondary_y=True)
data.groupby("Region").get_group("TAS")["2005-06-01"]["WetBulbTemperature"].plot(ax=axes[1,0])
data.groupby("Region").get_group("TAS")["2005-06-01"]["TotalDemand"].plot(ax=axes[1,0], secondary_y=True)
data.groupby("Region").get_group("SA")["2005-06-01"]["WetBulbTemperature"].plot(ax=axes[1,1])
data.groupby("Region").get_group("SA")["2005-06-01"]["TotalDemand"].plot(ax=axes[1,1], secondary_y=True)
```
